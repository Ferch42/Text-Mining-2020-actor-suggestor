```python
import pandas as pd
```

```python
data = pd.read_csv('movie_data_total.csv')
```

```python
data
```

```python
data = data[data.summary_wiki!="['NoText']"]
```

```python
data = data.drop(data[data.summary_wiki == "NoSearch"].index)
```

```python
data
```

```python
# parsing the input into data structures
columns = ['summary_wiki', 'cast', 'info_json']
def text_parser(text):
    result = []
    try:
        r = eval(text)
        result = r
    except:
        result = [text]
    return result

for c in columns:
    data[c] = [text_parser(x) for x in data[c]]
```

```python
data = data.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'])

def parse_summary(summary):
    if len(summary)==1:
        return str(summary[0])
    else:
        return ' '.join([str(s) for s in summary])
data['summary_wiki'] = data['summary_wiki'].apply(parse_summary)
```

```python
data.to_csv('cleaned_movie_data.csv')
```
